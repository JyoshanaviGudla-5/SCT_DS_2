```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Make sure 'train.csv' is uploaded to the Colab environment or adjust the path.
df = pd.read_csv('/train.csv')
print(df.head())
print(df.info())
```

```python
average_age = df['Age'].mean()
df['Age'].fillna(average_age)
print(round(average_age, 2))
df.drop('Cabin', axis=1)
most_common_port = df['Embarked'].mode()[0]
df['Embarked'].fillna(most_common_port)
print(most_common_port)
print("\n---Check whether any missing values are there in the dataset")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("--- Survival Rate by Gender (%) ---")
survival_by_sex = df.groupby('Sex')['Survived'].mean() * 100
print(survival_by_sex)
print("--- Survival Rate by Passenger Class (%) ---")
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print(survival_by_class)
plt.figure(figsize=(8,5))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.xlabel('0 = Did not Survive, 1 = Survived')
plt.ylabel('Number of Passengers')
plt.legend(title='Gender')
plt.show()
```
